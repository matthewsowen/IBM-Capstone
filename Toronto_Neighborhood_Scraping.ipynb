<H3>First, I will import the necessary libraries

```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

<H3>Now I will retrieve the html from the Wikipedia page in text form and parse it using BeautifulSoup to retrieve just the text of the table with the information on the different Toronto neighborhoods.

```python
wiki = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
wiki_soup = BeautifulSoup(wiki, 'lxml')
```

```python
table = wiki_soup.find('table').tbody.text

table
```

<h3>I will now split the text to isolate the values of the individual cells and filter out blank values to generate a list containing the table values.

```python
split_table = table.split('\n')
```

```python
split_table1 = list(filter(None, split_table))
```

```python
split_table1
```

<H3>Now I will create a list of lists to organize the data back into a tabular format.

```python
split_table2 = [['PostalCode', 'Borough', 'Neighborhood']]
```

```python
count = 1

while count < (len(split_table1)/3):
    
    split_table2.append(split_table1[(3*count):(3*count+3)])
    
    count +=1

split_table2
```

<H3> Now I can use this list to create a pandas dataframe with the appropriate column names.

```python
column_names = split_table2.pop(0)

Tor_data = pd.DataFrame(split_table2, columns = column_names)
```

```python
Tor_data
```

<H3> I will then drop all rows with an unassigned Borough and replace unassigned neighborhoods with their borough name.

```python
Tor_data2 = Tor_data[Tor_data.Borough != 'Not assigned']
```

```python
Tor_data2.loc[Tor_data2['Neighborhood']=='Not assigned', 'Neighborhood'] = Tor_data2.loc[Tor_data2['Neighborhood']=='Not assigned', 'Borough']
```

```python
Tor_data2.head(10)
```

<H3> And now I will group the neighborhoods that share a borough, assign these groupings to the neighborhood column of the table, and, finally, drop the resulting duplicate rows.

```python
Tor_data2.loc[:,'Neighborhood'] = Tor_data2.groupby(['PostalCode','Borough'], as_index=False).transform(lambda x: ', '.join(x))
```

```python
Toronto_data = Tor_data2.drop_duplicates()

Toronto_data.head(10)
```
